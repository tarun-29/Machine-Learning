```python
import numpy as np
import pandas as pd
```

```python
column_name = ['user_id', "item_id", "rating", "timestamp"]
```

```python
df = pd.read_csv("u.data",sep="\t", names=column_name)
```

```python
df.head()
```

```python
movie_title = pd.read_csv("Movie_Id_Titles")
```

```python
movie_title.head()
```

```python
df = pd.merge(df, movie_title, on="item_id")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("white")
```

```python
rating = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
rating.head()
```

```python
rating["num of ratings"] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
moviemat = df.pivot_table(index="user_id", columns="title", values="rating")
```

```python
moviemat.head()
```

```python
rating.sort_values('num of ratings',ascending=False).head(10)
```

```python
# starwars_user_ratings = moviemat["Star Wars (1997)"]
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(rating['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(rating['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
